```python
while True:
    try:
        votos_candidato1 = int(input("Ingresa el número de votos para el candidato 1: "))
        votos_candidato2 = int(input("Ingresa el número de votos para el candidato 2: "))
    except ValueError:
        print("ERROR: Debes introducir un número entero. No se aceptan letras ni valores vacíos.")
        print("Por favor vuelve a intentarlo.\n")
        continue
    total_votos = votos_candidato1 + votos_candidato2
    if votos_candidato1 > votos_candidato2:
        print("El ganador es el candidato 1")
    elif votos_candidato2 > votos_candidato1:
        print("El ganador es el candidato 2")
    else:
        print("Empate, no se puede determinar ganador")
    print(f"El total de votos que fueron ingresados es: {total_votos}")
    while True:
        respuesta = input("¿Deseas continuar con una nueva votación? (s/n): ").strip().lower()
        if respuesta in ('s', 'si'):
            print()
            break
        elif respuesta in ('n', 'no'):
            print("Que tenga un excelente día.")
            break
        else:
            print("Respuesta no válida. Por favor ingresa 's' para sí o 'n' para no.")
    if respuesta in ('n', 'no'):
        break
```
